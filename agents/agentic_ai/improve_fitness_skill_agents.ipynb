```python
from langchain.agents import create_agent
from langchain.tools import tool
from langchain_openai import ChatOpenAI
```

```python
@tool(description="Track daily walking goal of 10,000 steps")
def walking_10000_steps() -> dict:
    action = "Walking 10000 steps daily"
    priority = 1
    expected_time = "30 minutes per sessions"
    dependency = []
    return {
        "action": "Walking 10,000 steps daily",
        "priority": 1,
        "expected_time": "30 minutes",
        "dependency": []
    }

@tool(description="Track strength training routine 3 times per week.")
def strength_training() -> dict:
    action = "Strenth training 3 days a week"
    priority = 2
    expected_time = "30 minutes per session"
    dependency = []
    return {
        "action": "Strength training 3x/week",
        "priority": 2,
        "expected_time": "30 minutes",
        "dependency": []
    }

@tool
def track_calories() -> dict:
    """Track daily calorie intake."""
    return {
        "action": "Track calories daily",
        "priority": 3,
        "expected_time": "10 minutes per day",
        "dependency": []
    }

@tool
def sleep_routine() -> dict:
    """Ensure 7-8 hours of sleep each night."""
    return {
        "action": "Sleep 7-8 hours nightly",
        "priority": 4,
        "expected_time": "7-8 hours per night",
        "dependency": []
    }


@tool
def weekly_review() -> dict:
    """Review weekly progress on all subgoals."""
    return {
        "action": "Weekly progress review",
        "priority": 5,
        "expected_time": "30 minutes weekly",
        "dependency": []
    }


llm = ChatOpenAI(
    base_url="http://localhost:1234/v1",
    api_key="not-needed",
    model="local-model",
    temperature=0.2,
)

agent = create_agent(
    model=llm,
    tools=[walking_10000_steps, strength_training, track_calories, sleep_routine, weekly_review]
)

```

```python
response = agent.invoke(
    {
        "messages": [
            {"role": "system", "content": "You are a smart assistant. Use the tools if necessary"},
            {"role": "user", "content": "Please create a weekly fat loss and strength training plan using all available tools, include daily habits and weekly review."}
        ]
    },
    stream=True  # <--- enables streaming
)
print(response["messages"][-1].content)
```
