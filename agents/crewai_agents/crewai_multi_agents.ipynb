```python
from crewai import Agent, Task, Crew, Process, LLM
from langchain_openai import ChatOpenAI
```

```python
# llm = ChatOpenAI(api_key="not-needed",
#                 base_url="http://localhost:1234/v1",
#                 model="mistral")

llm = LLM(model="custom_model",
          base_url="http://localhost:1234/v1",
          api_key="abcd")

nutritionist = Agent(
    role="Senior Nutritionist",
    goal='Create a personalized 7-day meal plan based on caloric needs',
    backstory="You are an expert in dietetics. You specialize inbalancing macronutrients and ensuring meals are practical yet healthy.",
    verbose=True,
    allow_delegation=False,
    llm=llm
)

fitness_coach = Agent(
    role="Fitness Coach",
    goal="Develop a workout routine that complements the nutrition intake",
    backstory="You focus on functional fitness. You ensures the user's energy levels from their diet are utilized efficiently in training",
    verbose=True,
    allow_delegation=False,
    llm=llm
)


# Defining the tasks
nutrition_task = Task(
    description="""Analyze a user who wants to loose 5kg. Create a meal plan for 2000 calories per day focusing on high protein""",
    expected_output="A structured 7-days meal plan with breakfast, Lunch and Dinner",
    agent=nutritionist
)

fitness_task = Task(
    description="""Based on the 2000 calorie high protein, create a 3 day week strength training routinel""",
    expected_output="A workout schedule including exercises, sets, and reps.",
    agent=fitness_coach
)

# Forming the health crew
health_crew = Crew(
    agents=[nutritionist, fitness_coach],
    tasks=[nutrition_task, fitness_task],
    process=Process.sequential
)
```

```python

# 4. Kickoff the process
result = health_crew.kickoff()

print("\n\n########################")
print("## FINAL HEALTH PLAN ##")
print("########################\n")
print(result)
```
