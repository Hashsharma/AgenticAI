```python
from crewai import Agent, Task, Crew, Process, LLM
llm = LLM(model="custom_model",
          base_url="http://localhost:1234/v1",
          api_key="abcd")

# 1. Manager with Guardrails
manager = Agent(
    role="Health Program Manager",
    goal="Coordinate the health plan and ensure 100% safety and accuracy.",
    backstory="You are a strict supervisor. You check every calorie and every rep.",
    allow_delegation=True,
    llm=llm,
    max_iter=5,          # Stops the manager from looping forever
    verbose=True
)

# 2. Nutritionist with Max Retries
nutritionist = Agent(
    role="Senior Nutritionist",
    goal="Create a 2000 calorie plan.",
    backstory="Dietetics expert.",
    llm=llm,
    max_retry_limit=3,   # If the manager rejects the plan, try again 3 times
    verbose=True
)

# 3. Fitness Coach
fitness_coach = Agent(
    role="Fitness Coach",
    goal="Complement the diet with workouts.",
    backstory="Functional trainer.",
    llm=llm,
    max_iter=10,
    verbose=True
)

# Defining the tasks
nutrition_task = Task(
    description="""Analyze a user who wants to loose 5kg. Create a meal plan for 2000 calories per day focusing on high protein""",
    expected_output="A structured 7-days meal plan with breakfast, Lunch and Dinner",
    agent=nutritionist
)

fitness_task = Task(
    description="""Based on the 2000 calorie high protein, create a 3 day week strength training routinel""",
    expected_output="A workout schedule including exercises, sets, and reps.",
    agent=fitness_coach
)

# 4. The Crew (The "Upgrade")
health_crew = Crew(
    agents=[nutritionist, fitness_coach],
    tasks=[nutrition_task, fitness_task],
    process=Process.hierarchical, # Manager takes control here
    manager_agent=manager,
    planning=True,                # Creates the roadmap first
    planning_llm=llm              # Use your custom API for planning too
)
```

```python

result = health_crew.kickoff()
```
