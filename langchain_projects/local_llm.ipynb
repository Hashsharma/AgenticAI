```python
import requests
import json

url = "http://localhost:1234/v1/chat/completions"

headers = {
    "Content-Type": "application/json"
}

text_to_analyze = """
Agentic AI allows systems to autonomously make decisions,
use tools, and execute tasks without constant human input.
"""

payload = {
    "model": "local-model",   # LM Studio ignores this but requires it
    "messages": [
        {
            "role": "system",
            "content": "You are a helpful assistant that summarizes text clearly."
        },
        {
            "role": "user",
            "content": f"Summarize the following text:\n{text_to_analyze}"
        }
    ],
    "temperature": 0.3
}

response = requests.post(url, headers=headers, data=json.dumps(payload))

result = response.json()
summary = result["choices"][0]["message"]["content"]

print("Summary:")
print(summary)
```
