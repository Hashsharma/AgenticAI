```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
import re

```

```python
section_pattern = r"(###\s+[0-9A-Z ]+\. [A-Z ]+)"

sections = re.split(section_pattern, text)
sections
```

```python
section_chunks = []
for i in range(1, len(sections), 2):
    header = sections[i].strip()
    content = sections[i+1].strip()

    section_chunks.append({"header": header, "content": content})

section_chunks
```

```python
clause_pattern = r"(\d+\.\d+(\([a-z]\))?)"
final_chunks = []

for sec in section_chunks:
    header = sec['header']
    content = sec['content']
    content.strip()
    clauses = re.split(clause_pattern, content)
    # clauses.remove('')
    # clauses.remove(None)

    merged_clauses = []
    for i in range(1, len(clauses), 3):
    
        clause_no = clauses[i]
        clause_text = clauses[i+2]
        

        merged_clauses.append(f'{clause_no} {clause_text}')


    text_splitter = RecursiveCharacterTextSplitter(
        separators=["\n\n", "\n", " ", ""],
        chunk_size=400,
        chunk_overlap=50
    )

    for clause in merged_clauses:
        chunks = text_splitter.split_text(clause)
        
        for chunk in chunks:

            final_chunks.append({"section": header, "chunks": chunk})

    
    final_chunks

    

final_chunks
```
